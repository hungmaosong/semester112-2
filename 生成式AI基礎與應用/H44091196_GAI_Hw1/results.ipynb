```python
import numpy as np

# 設定總共類別
c = 10
# 模擬輸出 logits
x = np.random.rand(c)

# TODO
print("練習 1：softmax 函數:")
print("logits = ",x)

def softmax(i):
  return np.exp(x[i]) / np.sum(np.exp(x))

P = [softmax(i) for i in range(c)]
print("p = ",P)
```

**練習 1：數值轉換(紫外線強度等級 在最後面!!!)**

```python
import pandas as pd
import numpy as np

print("練習1：數值轉換")

df = pd.read_csv('drive/MyDrive/Colab Notebooks/467410-2022-08.csv')
new_df = df #複製一份資料

# 把 uv 轉為int，並將無法轉換的值設為 NaN
uv_numeric = pd.to_numeric(df['日最高紫外線指數'], errors='coerce')

# 把 NaN 值改成 -1
uv_numeric.fillna(-1, inplace=True)

# 把 uv 轉為int
uv_int = uv_numeric.astype(int)

# 把 uv_int 轉為字串
uv_str = uv_int.astype(str)

uv_str[(uv_int >= 0) & (uv_int <= 2)] = "低"
uv_str[(uv_int >= 3) & (uv_int <= 5)] = "中"
uv_str[(uv_int >= 6) & (uv_int <= 7)] = "高"
uv_str[(uv_int >= 8) & (uv_int <= 10)] = "甚高"
uv_str[uv_int >= 11] = "極高"
uv_str[0] = "紫外線強度等級"
uv_str.name = "紫外線強度等級"

#新增一個欄位表示紫外線強度等級
new_df['紫外線強度等級'] = uv_str

print(new_df)

print()

#計算每個等級的出現次數
low = 0
medium = 0
high = 0
very_high = 0
extreme_high = 0
for i in range(uv_str.size):
  if(uv_str[i] == "低"):
    low += 1
  elif(uv_str[i] == "中"):
    medium += 1
  elif(uv_str[i] == "高"):
    high += 1
  elif(uv_str[i] == "甚高"):
    very_high += 1
  elif(uv_str[i] == "極高"):
    extreme_high += 1

print("計算每個等級的出現次數:")
print("低 = ",low)
print("中 = ",medium)
print("高 = ",high)
print("甚高 = ",very_high)
print("極高 = ",extreme_high)
```

 **練習2：條件篩選**

```python
import pandas as pd
import numpy as np

print("練習2：條件篩選")

df = pd.read_csv('drive/MyDrive/Colab Notebooks/467410-2022-08.csv')

# 把 rain 轉為float，並將無法轉換的值設為 NaN
rain_numeric = pd.to_numeric(df['降水量(mm)'], errors='coerce')
# 把 NaN 值改成 0
rain_numeric.fillna(0, inplace=True)
# 把 rain 轉為float
rain_float = rain_numeric.astype(float)
rain_float[0] = "降水量(mm)"

# 把 hour 轉為float，並將無法轉換的值設為 NaN
hour_numeric = pd.to_numeric(df['降水時數(hour)'], errors='coerce')
# 把 NaN 值改成 0
hour_numeric.fillna(0, inplace=True)
# 把 rain 轉為float
hour_float = hour_numeric.astype(float)
hour_float[0] = "降水時數(hour)"

#計算出降水強度（mm/hr）
strength = pd.Series(index=hour_float.index, name='降水強度（mm/hr)', dtype=object) #創一個空的series
strength[0] = "降水強度（mm/hr)"
for i in range(1,strength.size):
  if(hour_float[i] == 0): #避免除以0
    strength[i] = 0
  else:
    strength[i] = rain_float[i] / hour_float[i]

print(strength)
print()

#計算平均強度
mean = strength.loc[1:strength.size].mean()

print("平均強度 = ",mean)
print()

#印出降水強度大於平均強度的日期及其相關資訊
for i in range(1,strength.size):
  if(strength[i] > mean):
    print(f"8月{i}日")
    print(df.iloc[i])
    print()
```

**練習 1：折線圖**

```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

print("練習1：折線圖")

df = pd.read_csv('drive/MyDrive/Colab Notebooks/467410-2022-08.csv')

# 把 temperature 轉為float，並將無法轉換的值設為 NaN
temperature_numeric = pd.to_numeric(df['氣溫(℃)'], errors='coerce')
# 把 NaN 值改成 0
temperature_numeric.fillna(0, inplace=True)
# 把 temperature 轉為float
temperature_float = temperature_numeric.astype(float)
temperature_float[0] = "氣溫(℃)"

# 把 rain 轉為float，並將無法轉換的值設為 NaN
rain_numeric = pd.to_numeric(df['降水量(mm)'], errors='coerce')
# 把 NaN 值改成 0
rain_numeric.fillna(0, inplace=True)
# 把 rain 轉為float
rain_float = rain_numeric.astype(float)
rain_float[0] = "降水量(mm)"

#作圖
myplot = plt.plot

x = np.arange(1, 32)
y1 = temperature_float[1:32]
y2 = rain_float[1:32]

myplot(
    x,
    y1,
    color = 'red',
    label = 'temprature'
)
plt.xlabel('Date')
plt.ylabel('temprature(℃)')

plt.legend()
plt.show()

myplot(
    x,
    y2,
    color = 'blue',
    label = 'rain'
)
plt.xlabel('Date')
plt.ylabel('rainfall(mm)')

plt.legend()
plt.show()
```

**練習 2：雷達圖**

```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

print("練習2：雷達圖")

df = pd.read_csv('drive/MyDrive/Colab Notebooks/467410-2022-08.csv')

# 把 rate 轉為float，並將無法轉換的值設為 NaN
rate_numeric = pd.to_numeric(df['風速(m/s)'], errors='coerce')
# 把 NaN 值改成 0
rate_numeric.fillna(0, inplace=True)
# 把 rate 轉為float
rate_float = rate_numeric.astype(float)
rate_float[0] = "風速(m/s)"

# 把 direction 轉為float，並將無法轉換的值設為 NaN
direction_numeric = pd.to_numeric(df['風向(360degree)'], errors='coerce')
# 把 NaN 值改成 0
direction_numeric.fillna(0, inplace=True)
# 把 direction 轉為float
direction_float = direction_numeric.astype(float)
direction_float[0] = "風向(360degree)"

# 將風向分為四個區間
direction_interval = pd.cut(direction_float[1:32], bins=[0, 90, 180, 270, 360], labels=['0-90', '90-180', '180-270', '270-360'])

# 計算每個區間的平均風速
avg_rate = rate_float.groupby(direction_interval).mean()
print(avg_rate)

# 風向區間
directions = avg_rate.index
# 平均風速數據
values = avg_rate.values

# 確定雷達圖中的角度
angles = np.linspace(0, 2 * np.pi, len(directions), endpoint=False).tolist()

# 將第一個數據點重複，以閉合雷達圖
values = np.concatenate((values,[values[0]]))
angles += angles[:1]

# 繪製雷達圖
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='yellow', alpha=0.6)

# 設定半徑資訊
radius_ticks = [0, 1, 2, 3, 4]
radius_labels = ['0', '1', '2', '3', '4']
ax.set_yticks(radius_ticks)
ax.set_yticklabels(radius_labels)

ax.set_xticks(angles[:-1]) #設置雷達圖的角度刻度位置，angles[:-1]表示取angles列表中除了最後一個元素外的所有元素作為刻度位置。
ax.set_xticklabels(directions)
ax.set_title('Average Wind Speed by Direction Interval', size=20, color='blue')

plt.show()
```

**練習 1：改善決策樹分類模型**

```python
# 練習 Hint
import sklearn

# 處理資料時會用到的工具
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 匯入填補缺失值的工具
from sklearn.impute import SimpleImputer
# 匯入 Label Encoder
from sklearn.preprocessing import LabelEncoder
# 匯入決策樹模型
from sklearn.tree import DecisionTreeClassifier
# 匯入準確度計算工具
from sklearn.metrics import accuracy_score
# 匯入 train_test_split 工具
from sklearn.model_selection import train_test_split

print("練習1：改善決策樹分類模型")
print("作法一:使用超參數")

df = pd.read_csv('drive/MyDrive/Colab Notebooks/titanic/train.csv')

# 取出訓練資料需要分析的資料欄位
df_x = df[['Sex', 'Age', 'Fare']]
# 取出訓練資料的答案
df_y = df['Survived']

# 數值型態資料前處理
# 創造 imputer 並設定填補策略
imputer = SimpleImputer(strategy='median')
age = df_x['Age'].to_numpy().reshape(-1, 1)
# 根據資料學習需要填補的值
imputer.fit(age)
# 填補缺失值
df_x['Age'] = imputer.transform(age)

# 類別型態資料前處理
# 創造 Label Encoder
le = LabelEncoder()
# 給予每個類別一個數值
le.fit(df_x['Sex'])
# 轉換所有類別成為數值
df_x['Sex'] = le.transform(df_x['Sex'])

# 分割 train and test sets，random_state 固定為 1012
train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, train_size=0.8, random_state=1012)

# 創造決策樹模型
# model = DecisionTreeClassifier(random_state=1012)
model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=10,
    max_leaf_nodes=5 ** 10,
)
# 訓練決策樹模型
model.fit(train_x, train_y)

# 確認模型是否訓練成功
pred_train = model.predict(train_x)
# 計算準確度
train_acc = accuracy_score(train_y, pred_train)

# 輸出準確度
print('train accuracy: {}'.format(train_acc))

# 確認模型是否訓練成功
pred_test = model.predict(test_x)
# 計算準確度
test_acc = accuracy_score(test_y, pred_test)

# 輸出準確度
print('test accuracy: {}'.format(test_acc))
```
